# Predicting taxi fares fairness in New York City

### Importing necessary packages

```python
# Data manipulation, cleaning and transformation
import pandas as pd
import numpy as np
import h3

# Visualizations
from matplotlib import pyplot as plt
from itertools import product

# Machine learning
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
import lightgbm as lgb
import sklearn.metrics as metrics
from sklearn.metrics import classification_report,confusion_matrix, ConfusionMatrixDisplay, roc_auc_score
```

### Importing CSV file and preliminary data exploration

```python
raw_nyc = pd.read_csv('Projet_30636_A22_Donnee_Principale.csv', index_col=0)
# Noticing several columns with NaN values. Those will be handled after splitting the data to p^revent data leakage.
raw_nyc.info()
```

```python
# "passenger_count" has an outlier value (999). Erroneous values will be treated after the split to avoid data leakage.
raw_nyc.describe()
```

```python
# Dropping rows where the target variable "fare_amount" has NaN values. Those can't be used to fit the model.
raw_nyc = raw_nyc.dropna(subset=["fare_amount"])
raw_nyc.head()
```

### Splitting data into train and split

```python
# Removing "fare_amount" from x_train and x_test to keep only features used to build the model.
# Keeping only "fare_amount" for y_train and y_test as this is the target variable we are trying to predict.

x = raw_nyc.drop(["fare_amount"],axis=1)
y = pd.DataFrame(raw_nyc["fare_amount"])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=8)
```

### Data cleaning

```python
# Replacing outliers in "passenger_count" with x_train median for both x_train and x_test. >7 is used as it's considered the maximum taxi passenger capacity.
# Using the median avoid the risk of getting a value with decimals that wouldn't make sense.

x_train['passenger_count'] = x_train['passenger_count'].fillna(x_train['passenger_count'].median())
x_train.loc[x_train['passenger_count'] > 7, 'passenger_count'] = x_train['passenger_count'].median()
x_test['passenger_count'] = x_test['passenger_count'].fillna(x_train['passenger_count'].median())
x_test.loc[x_test['passenger_count'] > 7, 'passenger_count'] = x_train['passenger_count'].median()
```

```python
# Transforming "fare_amount" into a binary version "fairness_num".
# 1 if the value is below 10. We consider this fare range to be fair.
# 0 if the value is equal or above 10. We consider this fare range to be unfair.

def classes(column):
    if column < 10:
        return 1
    else:
        return 0

y_train['fairness_num'] = y_train["fare_amount"].apply(classes)
y_test['fairness_num'] = y_test["fare_amount"].apply(classes)
```

```python
# Keeping only the newly created binmary column for the purpose of the exercice.
y_train = y_train["fairness_num"]
y_test = y_test["fairness_num"]
```

### Creating meaningful and informative features for our model

```python
# Converting "pickup_datetime" into a datetime object.

x_train['pickup_datetime'] =  pd.to_datetime(x_train['pickup_datetime'], format='%Y-%m-%d %H:%M:%S %Z').replace('nan', pd.NA)
x_test['pickup_datetime'] =  pd.to_datetime(x_test['pickup_datetime'], format='%Y-%m-%d %H:%M:%S %Z').replace('nan', pd.NA)

# Creating "day_of_week" column.
# From 0 to 6. 0 being Sunday and 6 being Saturday.
# Replacing NaN values with x_train median.
x_train['day_of_week'] = x_train['pickup_datetime'].dt.strftime('%w')
x_test['day_of_week'] = x_test['pickup_datetime'].dt.strftime('%w')
x_train['day_of_week'] = x_train['day_of_week'].fillna(x_train['day_of_week'].mode()[0]).astype(int)
x_test['day_of_week'] = x_test['day_of_week'].fillna(x_train['day_of_week'].mode()[0]).astype(int)

# Creating "AM/PM" column.
# 0 if AM and 1 if PM.
# Replacing NaN values with x_train median.
x_train['AM/PM'] = x_train['pickup_datetime'].dt.strftime('%H')
x_train['AM/PM'] = x_train['AM/PM'].fillna(x_train['AM/PM'].mode()[0]).astype(int)
x_train.loc[x_train["AM/PM"] <= 12, "AM/PM"] = 0
x_train.loc[x_train["AM/PM"]  >= 13, "AM/PM"] = 1
x_train['AM/PM'] = x_train['AM/PM']

x_test['AM/PM'] = x_test['pickup_datetime'].dt.strftime('%H')
x_test['AM/PM'] = x_test['AM/PM'].fillna(x_train['AM/PM'].mode()[0]).astype(int)
x_test.loc[x_test["AM/PM"] <= 12, "AM/PM"] = 0
x_test.loc[x_test["AM/PM"] >= 13, "AM/PM"] = 1
x_test['AM/PM'] = x_test['AM/PM']

# Creating a function to generate a column for week and weekend "wk_wknd".

def wkorwknd(series):
    if series == 0 or series == 6:
        return "0"
    else:
        return "1"
    
# 0 if weekend and 1 if weekday    
x_train['wk_wknd'] = x_train['day_of_week'].apply(wkorwknd).astype(int)
x_test['wk_wknd'] = x_test['day_of_week'].apply(wkorwknd).astype(int)

# Creating "time_of_day" column
# Replacing NaN values with x_train median.
x_train['time_of_day'] = x_train['pickup_datetime'].dt.strftime('%H')
x_train['time_of_day'] = x_train['time_of_day'].fillna(x_train['time_of_day'].mode()[0]).astype(int)
x_test['time_of_day'] = x_test['pickup_datetime'].dt.strftime('%H')
x_test['time_of_day'] = x_test['time_of_day'].fillna(x_train['time_of_day'].mode()[0]).astype(int)
```

```python
# Transforming latitudes and longitudes into distances using h3 package.
x_train['Distance'] = x_train.apply(lambda x_train: h3.point_dist((x_train['pickup_latitude'], x_train['pickup_longitude']), (x_train['dropoff_latitude'], x_train['dropoff_longitude']), unit='km'), axis=1)
x_test['Distance'] = x_test.apply(lambda x_test: h3.point_dist((x_test['pickup_latitude'], x_test['pickup_longitude']), (x_test['dropoff_latitude'], x_test['dropoff_longitude']), unit='km'), axis=1)
x_train.head()
```

```python
# Keeping only 2 decimals for each latitude and longitude so that our model is able to represent New York City's difference neighborhoods.
# Making sure not to perform any rounding as this would skew the data.
def stripgeo(series):
    series = [float('{0:.2f}'.format(x)) for x in series]
    return series
    
x_train["pickup_longitude"] = stripgeo(x_train["pickup_longitude"])
x_test["pickup_longitude"] = stripgeo(x_test["pickup_longitude"])
x_train["pickup_latitude"] = stripgeo(x_train["pickup_latitude"])
x_test["pickup_latitude"] = stripgeo(x_test["pickup_latitude"])

x_train["dropoff_longitude"] = stripgeo(x_train["dropoff_longitude"])
x_test["dropoff_longitude"] = stripgeo(x_test["dropoff_longitude"])
x_train["dropoff_latitude"] = stripgeo(x_train["dropoff_latitude"])
x_test["dropoff_latitude"] = stripgeo(x_test["dropoff_latitude"])

x_train.head()
```

```python
# Now that meaningful features have been created, we can put asside original features as they won't be used in the model. It would be redundantant to do so.
# Only "passenger count" is retained from the original features as it's the only one that hasn't been altered to create new features.

x_train.drop(["pickup_datetime"], axis=1, inplace=True)
x_test.drop(["pickup_datetime"], axis=1, inplace=True)
x_train.head()
```

### Decision Tree - Train

```python
# Finding the ideal depth for the decision tree.

parameters = {'max_depth':range(1,30)}
dtree = GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4)
dtree.fit(x_train, y_train)
tree_model = dtree.best_estimator_
print (dtree.best_score_, dtree.best_params_) 
```

```python
# Fitting decision tree.

dtree = DecisionTreeClassifier(random_state=6, max_depth=8, min_samples_split=600)
dtree.fit(x_train,y_train)
```

```python
# Variable relative importance. Distance is by far the most informative for the decision tree.
feat_importances_tree = pd.Series(dtree.feature_importances_, index=x_train.columns)
feat_importances_tree.nlargest(20).plot(kind='barh')
```

```python
# Tree visiualization sample.

feature_names = x_train.columns
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtree,
                   max_depth=3,
                   feature_names=feature_names,  
                   class_names={0:"Unfair", 1:"Fair"},
                   filled=True,
                   fontsize=12)
```

### Decision  Tree - Predictions

```python
predictions_tree = dtree.predict(x_test)
```

```python
# Comparing accuracy on train and test to see if underfitting or overfitting is occuring.
print('Training set score: {:.4f}'.format(dtree.score(x_train, y_train)))
print('Test set score: {:.4f}'.format(dtree.score(x_test, y_test)))
```

```python
print(classification_report(y_test,predictions_tree))
```

```python
# Confusion matrix.
cm = confusion_matrix(y_test,predictions_tree)
disp_cm = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtree.classes_)
disp_cm.plot()
plt.show()
```

```python
# Visualization of AUC ROC pour the decision tree.

probs_tree = dtree.predict_proba(x_test)
preds_tree = probs_tree[:,1]
fpr_tree, tpr_tree, threshold_tree = metrics.roc_curve(y_test, preds_tree)
roc_auc_tree = metrics.auc(fpr_tree, tpr_tree)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr_tree, tpr_tree, label = 'AUC = %0.4f' % roc_auc_tree)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
# Function that will be used to visualize lift for all models.

def plot_Lift_curve(y_val, y_pred, step=0.01):
    
    aux_lift = pd.DataFrame()
    aux_lift['real'] = y_val
    aux_lift['predicted'] = y_pred
    aux_lift.sort_values('predicted',ascending=False,inplace=True)

    x_val = np.arange(step,1+step,step)
    ratio_ones = aux_lift['real'].sum() / len(aux_lift)
    y_v = []
    
    for x in x_val:
        num_data = int(np.ceil(x*len(aux_lift)))
        data_here = aux_lift.iloc[:num_data,:]
        ratio_ones_here = data_here['real'].sum()/len(data_here)
        y_v.append(ratio_ones_here / ratio_ones)
           
    fig, axis = plt.subplots()
    fig.figsize = (40,40)
    axis.plot(x_val, y_v, 'g-', linewidth = 3, markersize = 5)
    axis.plot(x_val, np.ones(len(x_val)), 'k-')
    axis.set_xlabel('Proportion of sample')
    axis.set_ylabel('Lift')
    plt.title('Lift Curve')
    plt.show()
```

```python
# Lift decision tree.
dtree_pred = dtree.predict_proba(x_test)
plot_Lift_curve(y_test, dtree_pred[:,1])
```

```python
# Loop to help us find the best hyperparameters for the selected model (LightGBM)
range_estimator= range(100,1000,200)
range_leaves= range(20,100,15)
for i, n in product(range_estimator,range_leaves):
    ltgbm2 = lgb.LGBMClassifier(num_leaves=i,n_estimators=n)
    ltgbm2.fit(x_train, y_train)
    auc=roc_auc_score(y_test, ltgbm2.predict_proba(x_test)[:, 1])    
    print("For an estimator ={}, and a number of leaves of {}, auc is {:.4f}".format(i,n,auc))
```
